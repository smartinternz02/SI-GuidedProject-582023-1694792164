```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns
```

Univariate Analysis


```python
df= pd.read_csv('/House Price India - House Price India.csv.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Histogram
plt.figure(figsize=(8, 6))
sns.histplot(df['bedrooms'], kde=True, bins=30)
plt.title('Histogram of bedrooms')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar plot for a categorical variable
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='bedrooms')
plt.title('Count of each category in bedrooms')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Box plot for a numerical variable
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, y='livingarea')
plt.title('Box plot of Living Area')
plt.ylabel('Value')
plt.show()
```

Bi-Variate and MultiVariate Analysis

```python
#ScatterPlot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='condition', y='views', data=df)
plt.title('Scatterplot between condition and views')
plt.show()
```

```python
#Joint Plot
sns.jointplot(x= 'condition',y ='views',data=df)
```
